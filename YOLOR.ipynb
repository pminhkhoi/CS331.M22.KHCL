## Rename Images

```python
import os
```

```python
image_dir = '/content/drive/MyDrive/yolor/dataset/test/images'
label_dir = '/content/drive/MyDrive/yolor/dataset/test/labels'
```

```python
i = 0
for file in os.listdir(image_dir):
  image_name = os.path.join(image_dir, file)
  label_name = image_name.replace('.jpg', '.txt').replace('images', 'labels')
  print(image_name)
  print(label_name)
  x = str(i)
  while len(x) < 4:
    x = '0'+ x
  x = 'IMG_'+x
  new_image_name = '.'.join([x, file.split('.')[-1]])
  print(os.path.join(image_dir, new_image_name))
  new_label_name = new_image_name.replace('jpg', 'txt')
  print(os.path.join(label_dir, new_label_name))
  os.rename(image_name, os.path.join(image_dir, new_image_name))
  os.rename(label_name, os.path.join(label_dir, new_label_name))
  i+=1
```

## Make dataset file

```python
image_folder = '/content/drive/MyDrive/yolor/dataset/test/images'
des = '/content/drive/MyDrive/yolor/dataset'
```

```python
with open(os.path.join(des, 'test.txt'), 'w') as f:
  content = []
  for file in os.listdir(image_folder):
    content.append(image_folder+'/'+file+'\n')
  content = ''.join(content)
  f.write(content)
f.close()
```
